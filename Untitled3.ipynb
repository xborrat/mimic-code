```python
specialties = pd.read_gbq(
    '''
    SELECT specialty, COUNT(*) AS number_of_admissions
    FROM admissions
    GROUP BY specialty
    ORDER BY number_of_admissions DESC 
    '''
    , configuration=config_gbq)
specialties.plot(kind='bar', x='specialty')
```
